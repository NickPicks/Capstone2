```python
colorado = city_data[city_data.State == 'CO']
```

```python
#Came back up here to change the index name for easy merge
colorado = colorado.rename(columns={'RegionName': 'City'})
```

Time to decide which cities I want to focus on.  Grabbing data data from Wikipedia to get population 
#and elevation of cities.  The elevation variable will give a easy cut off to narrow df to just mountian towns

```python
url = "https://en.wikipedia.org/wiki/List_of_cities_in_the_United_States_by_elevation"
co_cities = pd.read_html(url)
```

!!Could use some help on this one.  How can I writh is to avoid redundency 

```python
#cities = pd.DataFrame([t for lst in co_cities[:] for t in lst])
cities9 = pd.DataFrame(co_cities[0])
cities8 = pd.DataFrame(co_cities[1])
cities7 = pd.DataFrame(co_cities[2])
cities6 = pd.DataFrame(co_cities[3])
```

```python
cities = cities9.append(cities8).append(cities7).append(cities6)
```

```python
cities
```

```python
cities = cities.drop(cities[['State', 'Average Yearly Temp']], axis=1)
```

```python
cities.head()
```

Time to merge price df and elevation df.
 

```python
df_merged = cities.merge(colorado, how='outer', on='City')
```

```python
df_merged
```

```python
#Getting rid of all cities with a lower elevation
df = df_merged[df_merged['Elevation'].notna()]
```

```python
df.describe()
```

```python
#df.apply(lambda x: x.count(), axis=1)
df_null = df.T.isnull().sum(axis=0).sort_values(ascending=False)
```

```python
df_null
```

```python
df = df[df.isnull().sum(axis=1) < 150]
```
